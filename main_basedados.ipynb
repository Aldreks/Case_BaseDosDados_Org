```python

### VALOR TOTAL PRODUO CAFE POR ANO

#Prepara dados
dftmp = df4.groupby(["ano"])["valor_total_producao_cafe"].sum() 

# CONFIG GRAFICO
plt.figure(figsize=(14, 7), facecolor=AzulFayete)   
ax = plt.axes() 
ax.set_facecolor(AzulNoturno)
# Cria Grafico
dftmp.plot.bar(x="ano", y="valor_total_producao_cafe", color=Aquamarine)
#Cor da escala dos eixos
plt.tick_params(axis="x", colors=Aquamarine, size=14)
plt.tick_params(axis="y", colors=Aquamarine, size=14)
plt.xticks(rotation='horizontal') 
plt.title('ANLISE VALOR TOTAL DE PRODUO DO CAF (1985, 1996, 2006, 2017)', fontdict={'family': 'serif','color' : 'SkyBlue','weight': 'bold','size': 14})
plt.ylim(0, 200000000) #escala de apresentao eixo Y
plt.xlabel('Ano', size = 14, fontdict={'weight': 'bold'}, color='SkyBlue')
plt.ylabel('Valor_Total_R$ (x100Mi)', size = 14, fontdict={'weight': 'bold'}, color='SkyBlue')
plt.show()
```

```python
from modules.postgres import Conector_postgres
```

```python
# Conecta ao banco Postgres
bd = Conector_postgres(host="127.0.0.1", db="BD_Censo_Agropecuario")
```

```python

## Valor Total da Produo de Caf/Estado (R$)

valor_total_prod_cafe_estado = bd.selecionar(
        f"SELECT sigla_uf, SUM(valor_total_producao_cafe) AS valor_total FROM MUNICIPIO GROUP BY sigla_uf ORDER BY sigla_uf;")

for i in valor_total_prod_cafe_estado:     
    print(f"Estado.............: {i[0]}")
    print(f"Total Produo Caf: R$ {i[1]}")
```

```python

## Total da Produo de Caf/Estado (toneladas)

total_prod_cafe_estado = bd.selecionar(
        f"SELECT sigla_uf, SUM(producao_cafe) AS total_prod_cafe FROM MUNICIPIO GROUP BY sigla_uf ORDER BY sigla_uf;")

for i in total_prod_cafe_estado:     
    print(f"Estado.............: {i[0]}")
    print(f"Total Produo Caf: {i[1]} toneladas")
```

```python

## Total da Produo de Leite/Estado (litros)

total_prod_leite_estado = bd.selecionar(
        f"SELECT sigla_uf, SUM(producao_leite) AS total_prod_leite FROM MUNICIPIO GROUP BY sigla_uf ORDER BY sigla_uf;")

for i in total_prod_leite_estado:     
    print(f"Estado..............: {i[0]}")
    print(f"Total Produo Leite: {i[1]} litros")
```

```python

## Total de Registros por Ano

total_regist_por_ano = bd.selecionar(
        f"SELECT ano, COUNT(sigla_uf) FROM MUNICIPIO GROUP BY ano ORDER BY ano;")  
    
for i in total_regist_por_ano:
    print(f"Ano............: {i[0]}")
    print(f"Total Registros: {i[1]}")
```

```python

## Total da Produo de Caf SP + MG (toneladas)

total_producao_cafe_SP_MG = bd.selecionar(
        f"SELECT sigla_uf, SUM(producao_cafe) FROM MUNICIPIO GROUP BY sigla_uf  WHERE sigla_uf = 'SP' OR sigla_uf = 'MG' ;")

for i in total_producao_cafe_SP_MG:
    print(f"Estado................: {i[0]}")
    print(f"Total Produo de Caf: {i[1]} toneladas")
```
